# GNSS COverage by provider (New feature: for each hexbin: coverage = count gnss / (count points in hexbin) )

```python
#processing:
df = df_full_divided

df_coverage_calc_list = []

for prov in [32, 35, 37]:
    df_prov = df[df.provider == prov]

    hexbins = df_prov.hexbin.unique()
    df_GNSS = df_prov[df_prov.determination_position==1]

    count_in_hexbin_detPos1_list,count_in_hexbin_detPos3_list,count_in_hexbin_detPos4_list = [], [], []
    for hb in hexbins:
        df_temp = df_prov[ df_prov.hexbin == hb ]
        
        try:
            count_in_hexbin_detPos1 = int( df_temp[df_temp.determination_position == 1].count_in_hexbin )
        except:
            count_in_hexbin_detPos1 = 0
        try:
            count_in_hexbin_detPos3 = int( df_temp[df_temp.determination_position == 3].count_in_hexbin )
        except:
            count_in_hexbin_detPos3 = 0
        try:
            count_in_hexbin_detPos4 = int( df_temp[df_temp.determination_position == 4].count_in_hexbin )
        except:
            count_in_hexbin_detPos4 = 0

        count_in_hexbin_detPos1_list.append(count_in_hexbin_detPos1)
        count_in_hexbin_detPos3_list.append(count_in_hexbin_detPos3)
        count_in_hexbin_detPos4_list.append(count_in_hexbin_detPos4)
    
    df_coverage_calc = pd.DataFrame( {"hexbin": hexbins, \
    "count_in_hexbin_detPos1": count_in_hexbin_detPos1_list,
    "count_in_hexbin_detPos3": count_in_hexbin_detPos3_list,
    "count_in_hexbin_detPos4": count_in_hexbin_detPos4_list,
    } )
    df_coverage_calc["coverage"] = df_coverage_calc["count_in_hexbin_detPos1"] / \
        (df_coverage_calc["count_in_hexbin_detPos1"]+df_coverage_calc["count_in_hexbin_detPos3"]+df_coverage_calc["count_in_hexbin_detPos4"])

    df_coverage_calc["provider"] = np.ones(len(df_coverage_calc))*prov
    df_coverage_calc = df_coverage_calc[["hexbin", "coverage", "provider"]]


    df_coverage_calc_list.append(df_coverage_calc)

df_coverage = pd.concat(df_coverage_calc_list)
```
